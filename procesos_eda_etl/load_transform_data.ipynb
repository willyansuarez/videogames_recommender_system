### Tramiento del archivo user_reviews

```python
# carga del archivo user_reviews
path_file = "archivos_originales/user_reviews.json.gz"
# path_file = "/content/drive/MyDrive/datasets/steam_files/user_reviews.json.gz"
def read_data_json(path_file):
    with gzip.open(path_file, "rt", encoding="utf-8") as file:
        return [ast.literal_eval(line.strip()) for line in file]
```

```python
# lectura de la data
data = read_data_json()
```

```python
# construir el dataframe
df_user_reviews = pd.DataFrame(data)
```

```python
# ver y verificar la creaci√≥n del registro
df_user_reviews.head(2)
```

```python
# exportar a parquet
df_user_reviews = df_user_reviews.to_parquet("archivos_parquet/df_user_reviews.parquet")
```

```python
# carga del archivo para verificar
df_user_reviews = pd.read_parquet("archivos_parquet/df_user_reviews.parquet")
```

```python
# visualizar
df_user_reviews.head(2)
```

**Fin tratamiento archivo *user_reviews***
