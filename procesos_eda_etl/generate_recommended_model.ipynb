```python
# eliminar columnas innecesarias
df_model.drop([
       'title', 'url', 'release_date', 'reviews_url', 'discount_price', 'price',
       'early_access', 'metascore', 'developer', 'user_id', 'steam_id',
       'items', 'items_count', 'publisher'
    ], axis=1, inplace=True)
```

```python
# elimino nulos
df_model.dropna(inplace=True)
```

```python
# elimino duplicados
df_model.drop_duplicates(inplace=True)
```

```python
# visualizar algunos datos
df_model['genres']
```

```python
# Verificar si una fila contiene caracteres extra√±os o en idiomas diferentes al ingl√©s
# se puede hacer por columna, pero se consider√≥ mejor por filas
row_contains = df_model.apply(lambda row: row.str.contains('[^\x00-\x7F]').any(), axis=1)
```

```python
# veerifico la cantidad
row_contains[row_contains == True].count()
```

```python
# eliminar todas esas filas
df_model = df_model[~row_contains]
```

```python
# reset del √≠ndice
df_model.reset_index(drop=True, inplace=True)
```

```python
# compruebo tama√±o y datos
df_model.shape
df_model.head()
```

```python
# verificar como est√°n los valores en las columnas escogidas, al parecer son tipo listas
# de listas, al ver el tipo de datos es object, si se intenta ingresar al segundo elemento
# de la supuesta lista en caso de haber mas de un elemento, solo devuelve una letra
df_model['genres'][0]
```

```python
# aplico m√©todos de str por columna, replace y strip
df_model['genres'] = df_model['genres'].apply(lambda gene: [g.strip().replace("'", "").replace("[", "").replace("]", "").replace(" ", "") for g in gene.split(",")])
df_model['tags'] = df_model['tags'].apply(lambda gene: [g.strip().replace("'", "").replace("[", "").replace("]", "").replace(" ", "") for g in gene.split(",")])
df_model['specs'] = df_model['specs'].apply(lambda gene: [g.strip().replace("'", "").replace("[", "").replace("]", "").replace("-", "").replace(" ", "") for g in gene.split(",")])
```

```python
# uno las columnas en una sola
df_model["tags_2"] = df_model['genres'] + df_model['tags'] + df_model['specs']
```
